# reading the dataset to python environment

```python
import pandas as pd

path_to_dataset = r"C:\Users\Muhammed Ashik.A\Downloads\iris.xls"
df = pd.read_excel(path_to_dataset)
```

# Display the columns in the dataset

```python
print(df.columns)
```

# Calculate the mean of each column of the dataset

```python
mean_values = df.mean()
print(mean_values)
```

# Check for the null values present in the dataset

```python
null_values = df.isnull().sum()
print(null_values)
```

# Perform meaningful visualizations using the dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df)
plt.title('Pairplot of Iris dataset')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Classification', y='SL', data=df)
plt.title('Sepal Length by Species')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Classification', y='PW', data=df)
plt.title('Petal Width by Species')
plt.show()
```
